```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from keras.datasets import mnist

# Load MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Flatten images to vectors
X = x_train.reshape(-1, 28*28)

# Train GMM with 10 components
n_components = 10
gmm = GaussianMixture(n_components=n_components, covariance_type='full', max_iter=100)
gmm.fit(X)

# Generate new samples
n_samples = 10

# Predict labels for the generated samples
labels = gmm.predict(X[:n_samples])

# Reshape samples to images
samples = gmm.sample(n_samples)
samples = np.array(samples[0])
samples = samples.reshape(-1, 28, 28)

# Plot samples
fig, axs = plt.subplots(2, 5, figsize=(10, 5))
axs = axs.flatten()
for i in range(10):
    axs[i].imshow(samples[i], cmap='gray')
    axs[i].set_title(f"Digit {labels[i]}")
    axs[i].axis('off')

plt.show()
```
